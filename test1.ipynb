### Problem #6 (17 pts)

Find the solution to the following system of equation.
$$2x + 5y + 2z = -38 $$
$$3x - 2y + 4z = 17 $$
$$-6x + y - 7z = -12$$

<ol>
  <li> Transform the system equation into a matrix form; create your matrix A and your vector b. (6 pts)
  <li> Compute the determinand of the matrix to find possile solutions. (3 pts)
  <li> Compute the inverse of the matrix. (3 pts)
  <li> solve the system. (5 pts)
</ol>

```python
# Transform the system equation into a matrix form; create your matrix A and your vector b

A = np.array([[2,5,2],[3,-2,4],[-6,1,-7]])
print(A)

b = np.array([[-38],[17],[-12]])
print(b)

# Compute the determinant of the matrix to find possible solutions and Compute the inverse of the matrix

det_A = np.linalg.det(A)
print(det_A)

inv_A = np.linalg.inv(A)
print(inv_A)

# Solve the System

sol = inv_A.dot(b)
print(sol)
```
